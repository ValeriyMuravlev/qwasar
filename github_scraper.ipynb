```python
import requests as requests
from bs4 import BeautifulSoup as bs

def getting_url(url):
  response = requests.get(url) 

def extract(getting_url):
  soup = bs(response.text, 'html.parser')
  extracted = soup.find_all("article", class_="Box-row")
  return extracted

def transform(extract):
  main_info = []
  extracted = []
  rep_counter = 1
  for row in extracted:
    rep_creator = row.select_one('h1.h3.lh-condensed').text.split()[0]
    rep_name = row.select_one('h1.h3.lh-condensed').text.split()[2]
    try:
      rep_lang = row.select_one('span.d-inline-block.ml-0.mr-3').text.strip()
      #rep_lang = row.select_one('div.f6.color-fg-muted.mt-2').text.strip().split()[0]
    except:
      rep_lang = 'Language not specified'
    try:
      devs = row.select_one('img.avatar.mb-1.avatar-user')['alt'] 
    except:
      devs = 'Developers not found'
    try:
      #nmb_stars = row.select_one('div.f6.color-fg-muted.mt-2').text.strip().split()[1]
      nmb_stars = row.select_one('a.d-inline-block.mr-3').text.strip()
    except:
      nmb_stars = 'No stars'
    rep_text = row.select_one('p.col-9.color-fg-muted.my-1.pr-4').text.strip()
    main_info = main_info.append(f'| {rep_counter} | {rep_name} | {rep_creator} | {rep_lang} | {devs} | {nmb_stars} | {rep_text} |')
    #main_info.append(f'| {rep_counter} | Repository name: {rep_name} | Repository creator: {rep_creator} | Repository text: {rep_text} | ')
    #main_info.append({"â„–": rep_counter, "Repository name": rep_name, "Repository creator": rep_creator, "Repository text": rep_text})
    rep_counter += 1
  return main_info
  
#print(main_info)

def format(transform):
  for el in main_info:
    print(f'| â„– | Repository name | Repository creator | Repository language | Built by | Total likes | Repository text | \n{el}\n')
    #print(el, "\n")

def main():
  url = "https://github.com/trending"
  test_response = getting_url(url)
  extract_data = extract(test_response)
  transform_data = transform(extract_data)
  formating_data = format(transform_data)
  print(formating_data)

main()
```
